## Clone GitHub to Colab

```python
# Save GitHub Token, user.name and user.email in Google Drive
# Read the token file
token_path = '/content/drive/MyDrive/Colab-Github.txt'
with open(token_path, 'r') as file:
    commands = file.readlines()

for command in commands:
    command = command.strip()  # Remove any leading/trailing whitespace
    if command.strip().startswith('git'):
      !{command}
    else:
      exec(command.strip())
```

```python
# Clone GitHub
username = 'ken011001'
repo = 'Colab-GitHub'
!git clone https://{token}@github.com/{username}/{repo}
```
